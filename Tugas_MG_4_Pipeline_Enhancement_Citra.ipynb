```python
# ============================================
# PIPELINE ENHANCEMENT CITRA
# Underexposed - Overexposed - Uneven Lighting
# ============================================

# 1. Import Library
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from skimage.measure import shannon_entropy

plt.rcParams['figure.figsize'] = (12,6)

# ============================================
# 2. Upload 3 Images
# ============================================

print("Upload 3 citra:")
print("1. Underexposed (gelap)")
print("2. Overexposed (terlalu terang)")
print("3. Iluminasi tidak merata")

uploaded = files.upload()

# Mengambil nama file
file_names = list(uploaded.keys())

img1 = cv2.imread(file_names[0])
img2 = cv2.imread(file_names[1])
img3 = cv2.imread(file_names[2])

# Convert ke grayscale
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

images = {
    "Underexposed": img1,
    "Overexposed": img2,
    "Uneven Lighting": img3
}

# ============================================
# 3. Fungsi Menampilkan Citra + Histogram
# ============================================

def show_image_hist(image, title):

    plt.figure(figsize=(10,4))

    plt.subplot(1,2,1)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.hist(image.ravel(), bins=256, range=(0,255))
    plt.title("Histogram " + title)

    plt.show()


# ============================================
# 4. POINT PROCESSING
# ============================================

# Negative Transformation
def negative_transform(img):
    return 255 - img


# Log Transformation
def log_transform(img):

    c = 255 / np.log(1 + np.max(img))
    log = c * np.log(1 + img)

    return np.array(log, dtype=np.uint8)


# Gamma Correction
def gamma_correction(img, gamma):

    table = np.array([
        ((i / 255.0) ** gamma) * 255
        for i in np.arange(256)
    ]).astype("uint8")

    return cv2.LUT(img, table)


# ============================================
# 5. HISTOGRAM BASED ENHANCEMENT
# ============================================

# Manual Contrast Stretching
def contrast_stretch_manual(img):

    r_min = np.min(img)
    r_max = np.max(img)

    stretch = (img - r_min) * (255/(r_max - r_min))

    return stretch.astype(np.uint8)


# Automatic Contrast Stretching
def contrast_stretch_auto(img):

    p2, p98 = np.percentile(img,(2,98))

    stretched = cv2.normalize(img,None,0,255,cv2.NORM_MINMAX)

    return stretched


# Histogram Equalization
def hist_equalization(img):

    return cv2.equalizeHist(img)


# CLAHE
def clahe_enhancement(img):

    clahe = cv2.createCLAHE(
        clipLimit=2.0,
        tileGridSize=(8,8)
    )

    return clahe.apply(img)


# ============================================
# 6. METRIK EVALUASI
# ============================================

def contrast_ratio(img):

    return (np.max(img)-np.min(img))/(np.max(img)+np.min(img)+1e-5)


def snr(img):

    signal = np.mean(img)
    noise = np.std(img)

    return signal/(noise+1e-5)


def evaluate(img):

    return {
        "Contrast": contrast_ratio(img),
        "Entropy": shannon_entropy(img),
        "SNR": snr(img)
    }


# ============================================
# 7. PROSES PIPELINE UNTUK SETIAP CITRA
# ============================================

for name, img in images.items():

    print("\n==============================")
    print("PROCESSING :", name)
    print("==============================")

    show_image_hist(img,"Original")

    # POINT PROCESSING
    neg = negative_transform(img)
    log = log_transform(img)

    gamma1 = gamma_correction(img,0.5)
    gamma2 = gamma_correction(img,1.5)
    gamma3 = gamma_correction(img,2.5)

    # HISTOGRAM BASED
    cs_manual = contrast_stretch_manual(img)
    cs_auto = contrast_stretch_auto(img)

    hist_eq = hist_equalization(img)
    clahe = clahe_enhancement(img)

    results = {
        "Negative":neg,
        "Log":log,
        "Gamma 0.5":gamma1,
        "Gamma 1.5":gamma2,
        "Gamma 2.5":gamma3,
        "Contrast Stretch Manual":cs_manual,
        "Contrast Stretch Auto":cs_auto,
        "Histogram Equalization":hist_eq,
        "CLAHE":clahe
    }

    # ============================================
    # 8. VISUALISASI HASIL
    # ============================================

    for title, result in results.items():

        show_image_hist(result,title)

        metrics = evaluate(result)

        print(title,"Metrics")
        print("Contrast :",metrics["Contrast"])
        print("Entropy  :",metrics["Entropy"])
        print("SNR      :",metrics["SNR"])
        print()


# ============================================
# SELESAI
# ============================================

print("Pipeline enhancement selesai dijalankan")
```
